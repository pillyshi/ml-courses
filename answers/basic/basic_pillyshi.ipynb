```python
import numpy as np
import matplotlib.pyplot as plt
```

-1から1までを100等分したnumpy.ndarrayを生成し，変数`x`に代入せよ．

ヒント: `numpy.linspace`を使うと良い．

```python
x = np.linspace(-1, 1, 100)
```

$y = f(x) = x$とする．上でつくった$x$から$y$を生成し，グラフにプロットせよ．

ヒント: `plt.plot`を使うと良い

```python
y = x
plt.plot(x, y, 'b-')
```

標準正規分布に従う乱数を100個生成し，変数`e`に代入せよ．なお，乱数のシードはrandom_stateとして固定すること．

```python
random_state = 0
rnd = np.random.RandomState(random_state)
e = rnd.normal(0, 1, size=100)
```

`y_observed = y + e`とし，この意味を考えよ．

```python
y_observed = y + e
```

`e`を標準正規分布に従うノイズとみなすと，真の値`y`にノイズが混ざったのが，観測値`y_observed`

`y_observed`をscatterプロットせよ．

ヒント: `plt.scatter`を使うと良い

```python
plt.scatter(x, y_observed)
```

```python
# おまけ
# ノイズの強さをsigmaで表す
# 小さい方が，yに近く
# sigmaを色々変えて試すと良い
sigma = 0.25
y_observed = y + sigma * e
plt.scatter(x, y_observed)
```
